```python
#Ans1. 

class node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.start = None;
        
    #Inserting an element in last
    def insertAtLast(self, value):
        new_node = node(value)
        if (self.start==None):
            self.start = new_node;
        else:
            temp = self.start
            while temp.next!=None:
                temp=temp.next
            temp.next = new_node;
            
    #Approach 1 (Using extra data structure --> List)
    #Deleting duplicate elements in linked List --> O(N*N) ==> N for traversing the linked list * N for searching in list
    def deleteDuplicates(self):
        if self.start == None:
            print("There are no elements in the list")
        else:
            distinct_node = []
            current = self.start
            previous = self.start
            distinct_node.append(current.data)
            while current.next!=None:
                current=current.next
                if current.data in distinct_node:
                    previous.next = current.next
                    current = previous
                else:
                    previous = current
                    distinct_node.append(current.data)
                    
    #Approach 2 (Using Extra Data Structure --> Dict)                
    #Deleting duplicate elements in linked List --> O(N*1) ==> N for traversing the linked list * 1 for searching in Dict
    def deleteDuplicates2(self):
        if self.start == None:
            print("There are no elements in the list")
        else:
            distinct_node = {}
            current = self.start
            previous = self.start
            distinct_node[current.data] = None
            while current.next!=None:
                current=current.next
                if current.data in distinct_node:
                    previous.next = current.next
                    current = previous
                else:
                    previous = current
                    distinct_node[current.data] = None
                    
                    
    #Approach 3 (Without Using Extra Data Structure)                
    #Deleting duplicate elements in linked List --> O(N*N) ==> N for traversing the linked list * N for traversing
    #                                                          the link list for runner variables
    def deleteDuplicates3(self):
        if self.start == None:
            print("There are no elements in the list")
        else:
            current = self.start
            while current:
                runner = current
                # Check until runner.next is not None.
                while runner.next:
                    if current.data == runner.next.data:
                        runner.next = runner.next.next
                    else:
                        runner = runner.next
                current = current.next
                    
                
            
    #Defining function to print
    def viewList(self):
        if (self.start==None):
            print("There are no elements in the list")
        else:
            temp = self.start
            while temp.next!=None:
                print(temp.data)
                temp=temp.next
            print(temp.data)

            
            

mylinklist = LinkedList()
print('Inserting 3 values at last of link list')
mylinklist.insertAtLast(1)
mylinklist.insertAtLast(1)
mylinklist.insertAtLast(1)
mylinklist.insertAtLast(1)
mylinklist.insertAtLast(1)
mylinklist.insertAtLast(1)
mylinklist.insertAtLast(1)
mylinklist.insertAtLast(1)
mylinklist.insertAtLast(1)
mylinklist.viewList()
print()
mylinklist.deleteDuplicates3()
mylinklist.viewList()
```

```python
#Ans4.  
#Approach1. Create 2 link list 

class node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.start = None;
        
    #Inserting an element in last
    def insertAtLast(self, value):
        new_node = node(value)
        if (self.start==None):
            self.start = new_node;
        else:
            temp = self.start
            while temp.next!=None:
                temp=temp.next
            temp.next = new_node;
    
    
    #partitioning linkedlist based on element  --> O(N) ==> N for traversing the linked list 
    #Space Complexity --> O(N) --> Creating 2 linked list where the max combined length is O(N)
    def partitionAtK(self, llist, k):
        #Creating 2 linked list, before (which will store all the elements that are smaller than the k) and after(which
        #will store all the elements that are equal or greater than the k)
        beforeStart = None  #Initialize to zero
        beforeEnd = None   #Initialize to zero
        afterStart = None  #Initialize to zero
        afterEnd = None   #Initialize to zero
        
        
        if (llist.start==None):
            print("There are no elements in the list")
            
        temp = llist.start
        while temp!=None:
            if temp.data < k:
                #Insert node in the end of the before list
                if beforeStart == None:
                    beforeStart = temp
                    beforeEnd = beforeStart
                else:
                    beforeEnd.next = temp
                    beforeEnd = temp
            else:
                #Insert node in the end of the after list
                if afterStart == None:
                    afterStart = temp
                    afterEnd = afterStart
                else:
                    afterEnd.next = temp
                    afterEnd = temp
                
            temp = temp.next
            
        #Making the end of the linked list None    
        afterEnd.next = None
        
        #Returning the after linked list if the before linked list is none
        if beforeStart == None:
            return afterStart
        
        #Merging the before and after linked list
        beforeEnd.next = afterStart 
        
        return beforeStart
        
    
    
    #Defining function to print
    def viewList(self):
        if (self.start==None):
            print("There are no elements in the list")
        else:
            temp = self.start
            while temp.next!=None:
                print(temp.data)
                temp=temp.next
            print(temp.data)

   
            

mylinklist = LinkedList()
print('Inserting values at last of link list')
mylinklist.insertAtLast(1)
mylinklist.insertAtLast(10)
mylinklist.insertAtLast(6)
mylinklist.insertAtLast(3)
mylinklist.insertAtLast(2)
mylinklist.insertAtLast(1)
mylinklist.insertAtLast(7)
mylinklist.insertAtLast(5)
mylinklist.insertAtLast(3)
mylinklist.viewList()
print()
k = 2
partitionList = mylinklist.partitionAtK(mylinklist, k)

print("New Partition Linked List at:", k)
while partitionList.next!=None:
    print(partitionList.data)
    partitionList = partitionList.next
```

```python
#Ans6.  
#Approach1. Create 2 link list 

class node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.start = None;
        
    #Inserting an element in last
    def insertAtLast(self, value):
        new_node = node(value)
        if (self.start==None):
            self.start = new_node;
        else:
            temp = self.start
            while temp.next!=None:
                temp=temp.next
            temp.next = new_node;
    
    
    #Approach 1. Create a dictionary and store the values
    #Time Complexity -- O(N+N) => O(N)
    #Space Complexity -- O(N) as creating an extra data structure
    def ifPalindrome1(self):
        if self.start==None:
            print("Linked List is empty")
        
        pal_dict = {}
        temp = self.start
        while temp!=None:
            if temp.data in pal_dict:
                pal_dict[temp.data] += 1
            else:
                pal_dict[temp.data] = 1
                
            temp = temp.next
        
        odd_count = 0
        for key,val in pal_dict.items():
            if val%2!=0:
                odd_count+=1
                
        if odd_count>1:
            return False
        else:
            return True
                
    
    #Approach 2: Reverse the linked list and then compare
    #Time Complexity -- O(N+N/2) => O(N)
    #Space Complexity -- O(N) as creating an extra data structure - linked list
    def ifPalindrome2(self):
        if self.start==None:
            print("Linked List is empty")
        
        current = self.start
        prev = None
        llen = 1
        while current!=None:          #Using this code, a linked list will be created "Prev"
            #print(current.data)
            head = current.next
            current.next = prev
            prev = current
            current = head
            llen+=1
       
        #checking the reverse linked list and actual linked list
        t = 0
        temp = self.start
        while t<= llen/2:
            if temp.data != prev.data:
                return False
            t += 1
        return True
            
            
        
        
                
    #Defining function to print
    def viewList(self):
        if (self.start==None):
            print("There are no elements in the list")
        else:
            temp = self.start
            while temp.next!=None:
                print(temp.data)
                temp=temp.next
            print(temp.data)

   
            

mylinklist1 = LinkedList()
print('Inserting values at last of link list 1')
mylinklist1.insertAtLast(1)
mylinklist1.insertAtLast(2)
mylinklist1.insertAtLast(3)
mylinklist1.insertAtLast(4)
mylinklist1.insertAtLast(3)
mylinklist1.insertAtLast(1)
mylinklist1.viewList()

print("The linked list is a Palindrome:")
mylinklist1.ifPalindrome2()

#mylinklist1.viewList()
```
